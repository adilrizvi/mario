```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np
import datetime
```

```python
train = pd.read_csv('train.csv')
```

```python
train.describe(include='all')
```

Only few records are missing from `name`(1 record) and `desc`(8 records) column , hence we should drop these rows.

```python
null_name = train.loc[train.name.isnull(), :]
null_name.shape

null_desc = train.loc[train.desc.isnull(), :]
null_desc.shape

# dropping the null records for name and desc
train = train.loc[train.name.notnull(), :]
train = train.loc[train.desc.notnull(), :]
```

```python
train.describe(include='all')
```

```python
# converting all the unix time stamp to human readable for analysis into another copy
unix_time_data = train.copy(deep=True)

train.loc[:, 'deadline'] = train.deadline.apply(lambda x:
        datetime.datetime.fromtimestamp(
        x).strftime('%Y-%m-%d'))
    
train.loc[:, 'state_changed_at'] = train.state_changed_at.apply(lambda x:
        datetime.datetime.fromtimestamp(
        x).strftime('%Y-%m-%d'))
    
train.loc[:, 'created_at'] = train.created_at.apply(lambda x:
        datetime.datetime.fromtimestamp(
        x).strftime('%Y-%m-%d'))
    
train.loc[:, 'launched_at'] = train.launched_at.apply(lambda x:
        datetime.datetime.fromtimestamp(
        x).strftime('%Y-%m-%d'))
```
